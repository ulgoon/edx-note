### Project: Week 2 - Find your Data
프로젝트: 두번째 주 – 데이터 찾기
두번째 주: 여러분의 데이터를 찾으세요

지난 주에 다뤘던 기사 기획에 기반해서, 기사를 위해 필요한 데이터를 조사해 보세요.

여러분이 필요한 데이터를 한 시간 안에 찾지 못한다 해도, 그것은 전혀 이상한 일이 아닙니다! 항상 일어나는 일이죠. 너무 빨리 포기하지 마세요. 지리 (가령, 전세계에서 한 국가로) 또는 시간 (가령 10년 추세 분석에서 3년으로)을 좁혀보세요. 정보원(네, 사람이요)한테 전화를 걸어 보세요. 그리고 어디서 데이터를 얻을 수 있을지 물어보세요. 아니면 정보공개청구를 요청해 보세요.
여러분이 포럼에 제출해야 하는 것은:
1)   데이터를 수집했다는 것을 증명할 데이터 파일(들), 또는 스크린 샷. 온라인에서 찾았다면 링크도 포함시키세요. 프라이버시, 보안, 기밀, 단독 기사 등의 이유로 공개를 원하지 않는다면, 괜찮습니다. 어떤 종류의 데이터를 수집했는지, 어떤 형식이고, 어떤 단위이고, 등등의 정보를 설명하세요.
2)   한 문단으로 어디서 어떻게 데이터를 수집했는지 설명하세요. 누가 데이터를 수집하고 발행하는지에 대한 세부사항도 포함해서요. 데이터의 약점이 있다면 논의하세요. 예상치 못한 사건들이 과정에 방해가 됐거나 도움이 됐다면, 여러분의 경험 또한 공유하세요.

포럼에 제출 할 필요는 없지만 시도해 볼 만한 가치가 있는 것은: 자신의 경험과 생각을 기록하기 위한 데이터 로그를 시작하는 것입니다. 예를 들자면, 전에는NYTimes와 일했고 현재ProPublica에 있는 데이터 저널리스트인 Derek Willis의 데이터 로그가 있습니다: http://dwillis.github.io/data-reporting/data_log.pdf
